# Smarket Data:

Daily movements in S&P stock index 

A 5-year period between 2001 and 2005. 

# Reason for Analysis 

To predict whether the index will increase or decrease on a given day using the past 5 days’ percentage changes in the index. 

# Insights: 

Yesterday: displays two boxplots of the previous day’s percentage changes in the stock index
one for the 648 days for which the market increased on the subsequent day.
one for the 602 days for which the market decreased. 
The two plots look identical, hinting that there is no simple strategy for using yesterday’s movement in the S&P to predict today’s returns.

Two & Three Days Previous:
Also indicate little association between past and present returns.
This is expected as in the presence of strong correlations between successive days’ returns, you can adopt a simple trading strategy to generate profits from the market. 


# FIGURE(1.2)

```python
fig, ax = plt.subplots(1,3, figsize=(15, 7)) # range of charts # FIGURE 1.2
# we are comparing Tpday's Direction with the past previous thress days using a box and whisker
# sns.boxplot(x=Smarket["Today's Direction"], y=Smarket["Yesterday"], ax=ax[0])
# We need to swith the order of the "up" and "Down buckets"
sns.boxplot(x=Smarket["Today's Direction"], y=Smarket["Yesterday"], order=["Down","Up"], ax=ax[0])
# No that we have Yesterdays we want Two Days Previous and then Three Days Previous"
sns.boxplot(x=Smarket["Today's Direction"], y=Smarket["Two Days Previous"], order=["Down","Up"], ax=ax[1])
sns.boxplot(x=Smarket["Today's Direction"], y=Smarket["Three Days Previous"], order=["Down","Up"], ax=ax[2])
```

```python
# the last figure with this data set states that a quadratic discriminant analysis model was fitted to the subset
# of the Smarket data corresponding to the 2001–2004 time period, and predicted
# the probability of a stock market decrease using the 2005 data
# so using sklearn we are going to Apply the probaility
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
Smarket.head()
```

```python
# PRINT THE OUT SO YOU CAN SEE THE OUTPUT OF EACH
# We want to have our trainging set to have its X inputs for columns Day_1 and Day_2 FROM 2001 to 2004
X_train = Smarket[:'2004'][['Day_1','Day_2']]
# and the y inputs for the training model to be Today's direction from 2004
y_train = Smarket[:'2004']["Today's Direction"]
# then we want the X input for the testing set to be FROM JUST 2005 !!!
X_test = Smarket['2005':][['Day_1','Day_2']]
# and the y input for testing to be FROM JUST 2005 !!!
y_test = Smarket['2005':]["Today's Direction"]
# then we call the quadratic descritmant Analysis methond
qda = QuadraticDiscriminantAnalysis()
# and create a new variable called Pred for prediction and we will apply the QDA method to it
pred = qda.fit(X_train, y_train).predict(X_test)
```

```python
# this creates out prediction Probability
pred_p = qda.predict_proba(X_test)
```

```python
ProbabilityPercentage = pd.DataFrame(data=pred_p, columns=["Up", "Down"]) # Group
```

```python
ProbabilityPercentage.head()
```

# Insight:

In Chapter 4, we explore this data using several different statistical learning methods. 

Interestingly, there are hints of some weak trends in the data that suggest that, at least for this 5-year period, it is possible to correctly predict the direction of movement in the market around 60% of the time


# FIGURE(1.3)

```python
sns.boxplot(data=ProbabilityPercentage, order=["Down","Up"])  #FIGURE 1.3.
plt.xlabel("Today's Direction", fontsize=14) 
plt.ylabel("Predicted Probability", fontsize=14)
```

# Gene Data

NCI60 dataset consists of 6,830 gene expression measurements for each of 64 cancer cell lines.

# Reason for Analysis:

To cluster groups among the cell lines based on their gene expression measurements, in order to better understand the relationship between gene expression levels and cancer.

# Insight:

Left Panel

Representing each of the 64 cell lines using just two numbers, Z1 and Z2.

These are the first two principal components of the data.

They summarize the 6,830 expression measurements for each cell line down to two dimensions. 

There are at least four groups of cell lines,represented using separate colors. 

Then you can  examine the cell lines within each cluster for similarities in their types of cancer.

The cell lines correspond to 14 different types of cancer. 

Right Panel

Identical to the left-hand panel, except that the 14 cancer types are shown using distinct colored symbols (or more different colors). 
cell lines with the same cancer type tend to be located near each other in this two-dimensional representation. 


# FIGURE(1.4)

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,6)) # create the frame the the 2 plots

cancer_color = pd.factorize(y.type)[0] # tks the cancer types and color codes tham
cmap = plt.cm.tab10 # give the theme of the color cod

# Left plot
# ax1.scatter(NCI60_plot.iloc[:,0], NCI60_plot.iloc[:,1]) the fist thing we want to do is reverse the order of our y
# ax1.scatter(NCI60_plot.iloc[:,0], -NCI60_plot.iloc[:,1], s=50) lets make out dots bigger
# ax1.scatter(NCI60_plot.iloc[:,0], -NCI60_plot.iloc[:,1], cmap=cmap, s=50) the plot needs to know how to sort the colors
ax1.scatter(NCI60_plot.iloc[:,0], -NCI60_plot.iloc[:,1], c=kmeans.labels_, cmap=cmap, s=50) # c=kmeans.labels_ creates the groups
ax1.set_ylabel('Z2') # and note in the book the y axis is labeled as Z2 we will name the x axis in a momnet

# Right plot same thing but now we are labeling it with the Princapl component Analyis
ax2.scatter(NCI60_plot.iloc[:,0], -NCI60_plot.iloc[:,1], c=cancer_color, cmap=cmap, s=50)
ax2.set_ylabel('Z2')

# xlabel for both plots
for ax in fig.axes:
    ax.set_xlabel('Z1')
```
